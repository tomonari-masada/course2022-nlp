```
from transformers import BertJapaneseTokenizer

tokenizer = BertJapaneseTokenizer.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
```

```
text = (
    "岸田文雄首相は７日、米グーグルのピチャイ最高経営責任者（ＣＥＯ）と官邸で面会し、"
    "日本のデジタル化支援のため、令和６年までに１千億円を日本で投資する計画について説明を受けた。"
    "面会後、ピチャイ氏が官邸で記者団の取材に応じた。"
    "インフラや人材育成、リスキリング（学び直し）などに投資する。"
)
```

```
print('Original: ', text)
print('Tokenized: ', tokenizer.tokenize(text))
print('Token IDs: ', tokenizer.convert_tokens_to_ids(tokenizer.tokenize(text)))
```
